```python
#The goal of this project is the find out wish high school as the highest scoring of 80 or above through out the years
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
df.head()
```

```python
#Finding out the time frame of our dataset
df['Year'].unique()
```

```python
#According to the  Data Dictionary - Dataset Information file:
# "Pursuant to the legislation and in accordance with the Family Educational Rights and Privacy Act (FERPA), rows with 5 or fewer students have been replaced with an "s". In addition,
# other rows have been replaced with an "s" when they could reveal, through addition or subtraction, the underlying numbers that have been redacted..."
# For this reason I will be dropping these values
df.drop(df[df['Mean Score'] == 's'].index, inplace = True)
df.count()
```

```python
# dropping off some columns that i won't be needing
df.drop(['Mean Score','Number Scoring Below 65','Number Scoring 65 or Above','Number Scoring 80 or Above','Number Scoring CR','Percent Scoring CR'], axis=1, inplace = True)
df.head()
```

```python
#In this project i will focusing on:
#school Level: High school
#Regents Exam : Comon Core Algebra
#Year: 2015-2016-2017
#Percent Scoring 80 or Above
df_2015=df[(df['Year']==2015)&(df['School Level']=='High school')&(df['Regents Exam']=='Common Core Algebra')]
df_2015.head()
```

```python
df_2015.isnull().sum()
```

```python
#I will focus on Total Tested tha is not less than 30 since 30 is is often used as a rule of thumb for a minimum sample size in statistics because it is the point at which the central limit theorem begins to apply
#Top 10 spots
df_2015=df_2015[df_2015['Total Tested']>30].sort_values(by='Percent Scoring 80 or Above',ascending=False).head(10)
df_2015
```

```python
df_2015['Percent Scoring 80 or Above']=df_2015['Percent Scoring 80 or Above'].astype(float)
df_2015.plot(x='School Name',y='Percent Scoring 80 or Above',kind='bar')
```

```python
df_2016=df[(df['Year']==2016)&(df['School Level']=='High school')&(df['Regents Exam']=='Common Core Algebra')]
df_2016.head()
```

```python
df_2016.isnull().sum()
```

```python
df_2016=df_2016[df_2016['Total Tested']>30].sort_values(by='Percent Scoring 80 or Above',ascending=False).head(10)
df_2016
```

```python
df_2016['Percent Scoring 80 or Above']=df_2016['Percent Scoring 80 or Above'].astype(float)
df_2016.plot(x='School Name',y='Percent Scoring 80 or Above',kind='bar')
```
